```python
#Funcion para tranformar los labels a formato legible por el Modelo YOLOV - 1

import json

# Ruta del archivo JSON
json_file_path = r"\\wsl.localhost\Ubuntu-22.04\home\santiago_e\Object-detection\via_project(train)_23Mar2023_20h12m_json.json"

# Ruta donde se guardarán los archivos de texto con las coordenadas de la caja delimitadora
output_dir = "data_set\test"

# Nombre de las clases y sus respectivos índices
class_names = {"car": 0, "bus": 1, "motocicle": 2, "street sign": 3, "building": 4, "people": 5}

with open(json_file_path, "r") as f:
    data = json.load(f)

for img_key in data:
    # Obtener el nombre del archivo y tamaño de la imagen
    img_filename = data[img_key]["filename"]
    img_size = data[img_key]["size"]
    
    # Obtener las regiones de la imagen
    regions = data[img_key]["regions"]
    
    # Crear el archivo de salida para la imagen actual
    output_file_path = output_dir + img_filename.replace(".jpg", ".txt")
    
    with open(output_file_path, "w") as f:
        for region in regions:
            # Obtener las coordenadas del polígono
            all_points_x = region["shape_attributes"]["all_points_x"]
            all_points_y = region["shape_attributes"]["all_points_y"]
            
            # Calcular las coordenadas de la caja delimitadora
            x_min = min(all_points_x)
            x_max = max(all_points_x)
            y_min = min(all_points_y)
            y_max = max(all_points_y)
            
            # Calcular las coordenadas normalizadas
            x_center = (x_min + x_max) / 2 / img_size
            y_center = (y_min + y_max) / 2 / img_size
            width = (x_max - x_min) / img_size
            height = (y_max - y_min) / img_size
            
            # Obtener la clase del objeto
            object_class = list(region["region_attributes"].keys())[0]
            class_index = class_names[object_class]
            
            # Escribir las coordenadas de la caja delimitadora en el archivo de salida
            f.write(f"{class_index} {x_center} {y_center} {width} {height}\n")
```

```python
#Funcion para tranformar los labels a formato legible por el Modelo YOLOV - 2


# Ruta del archivo JSON
json_file_path = r"\\wsl.localhost\Ubuntu-22.04\home\santiago_e\Object-detection\via_project(train)_23Mar2023_20h12m_json.json"

# Ruta donde se guardarán los archivos de texto con las coordenadas de la caja delimitadora
output_dir = "data_set\test"

# Nombre de las clases y sus respectivos índices
class_names = {"car": 0, "bus": 1, "motocicle": 2, "street sign": 3, "building": 4, "people": 5}

with open(json_file_path, "r") as f:
    data = json.load(f)

for img_key in data:
    # Obtener el nombre del archivo y tamaño de la imagen
    img_filename = data[img_key]["filename"]
    img_size = data[img_key]["size"]
    
    # Obtener las regiones de la imagen
    regions = data[img_key]["regions"]
    
    # Crear el archivo de salida para la imagen actual
    output_file_path = output_dir + img_filename.replace(".jpg", ".txt")
    
    with open(output_file_path, "w") as f:
        for region in regions:
            shape_attributes = region["shape_attributes"]
            region_attributes = region["region_attributes"]
            
            # Comprobar que se han definido las coordenadas de la caja delimitadora
            if "x" in shape_attributes:
                x_min = shape_attributes["x"]
                y_min = shape_attributes["y"]
                x_max = x_min + shape_attributes["width"]
                y_max = y_min + shape_attributes["height"]
            elif "all_points_x" in shape_attributes:
                all_points_x = shape_attributes["all_points_x"]
                all_points_y = shape_attributes["all_points_y"]
                x_min = min(all_points_x)
                x_max = max(all_points_x)
                y_min = min(all_points_y)
                y_max = max(all_points_y)
            else:
                continue  # saltar esta región si no se pueden obtener las coordenadas de la caja delimitadora
            
            # Calcular las coordenadas normalizadas
            x_center = (x_min + x_max) / 2 / img_size
            y_center = (y_min + y_max) / 2 / img_size
            width = (x_max - x_min) / img_size
            height = (y_max - y_min) / img_size
            
            # Obtener la clase del objeto
            object_class = list(region_attributes.keys())[0]
            class_index = class_names[object_class]
            
            # Escribir las coordenadas de la caja delimitadora en el archivo de salida
            f.write(f"{class_index} {x_center} {y_center} {width} {height}\n")
```
